## **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

#Dataframe dari list hasil

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

#Tampilkan dataframe

display(results_df)
```

## **Comparison(Perbandingan)**

```python
perbandingan_region = df.groupby('region')['climate_risk_score'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))

sns.barplot(x=perbandingan_region.index, y=perbandingan_region.values, palette='viridis', hue=perbandingan_region.index)
plt.title('Rata-rata Climate Risk Score Berdasarkan Region')
plt.xlabel('region')
plt.ylabel('Rata-rata Climate Risk Score')
plt.xticks(rotation=45)
plt.show()
```

**Kesimpulan:**

```python
risk_by_region = df.groupby('region')['climate_risk_score'].mean().sort_values(ascending=False)
risk_by_region_top3 = risk_by_region.head(3)

plt.figure(figsize=(8, 6))
risk_by_region_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Region dengan Climate Risk Score Tertinggi')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

```python
# Identifikasi 5 negara teratas berdasarkan jumlah data
top5negara = df['country'].value_counts().head(5).index
# Filter dataframe
df_top5_countries = df[df['country'].isin(top5negara)]
plt.figure(figsize=(12,7))
df_top5_countries.boxplot(
    column='climate_risk_score',
    by='country',
    ax=plt.gca(),
    grid=False,
    patch_artist=True,
    medianprops=dict(color='red')
)
plt.title('Distribusi Climate Risk Score di Top 5 Negara', fontsize=16)
plt.xlabel('Negara', fontsize=12)
plt.ylabel('Climate Risk Score', fontsize=12)

plt.xticks(rotation=45, ha='right')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

Grafik boxplot menampilkan **distribusi nilai *climate risk score* pada lima negara dengan jumlah data terbanyak** dalam dataset, sehingga memperlihatkan perbandingan sebaran, median, serta variasi nilai risiko iklim pada masing-masing negara.

**Insight:** Perbedaan posisi median dan lebar boxplot menunjukkan bahwa setiap negara memiliki tingkat risiko iklim dan variasi data yang berbeda, di mana negara dengan median lebih tinggi atau sebaran lebih luas cenderung memiliki tingkat kerentanan iklim yang lebih besar.

**Action:** Negara dengan nilai median risiko tinggi atau variasi yang besar dapat menjadi fokus dalam analisis lanjutan dan perumusan strategi mitigasi serta adaptasi perubahan iklim untuk mengurangi potensi dampak yang lebih besar.

# **Bubble Chart**

```python
plt.figure(figsize=(10,8))
sns.scatterplot(
    x='co2_emissions_mt',
    y='temperature_change_c',
    size='population_affected_m',
    hue='region',
    data=df,
    sizes=(50,1000),
    alpha=0.7,
    palette='viridis'
)
plt.title('Emisi CO₂ vs Perubahan Suhu (Ukuran: Populasi Terdampak, Warna: Region)', fontsize=16)
plt.xlabel('Emisi CO₂ (Mt)', fontsize=12)
plt.ylabel('Perubahan Suhu (°C)', fontsize=12)
plt.grid(True)
plt.legend(title='Region', bbox_to_anchor=(1.05,1), loc='upper left')

plt.tight_layout()
plt.show()
```

Grafik tersebut merupakan **scatter/bubble plot yang menunjukkan hubungan antara emisi CO₂ dan perubahan suhu**, di mana ukuran titik merepresentasikan **jumlah populasi terdampak** dan warna menunjukkan **wilayah (region)**.

**Insight:** Sebaran titik menunjukkan bahwa perubahan suhu terjadi pada berbagai tingkat emisi CO₂ di berbagai wilayah. Beberapa titik dengan ukuran besar menandakan bahwa peningkatan suhu dapat berdampak pada populasi yang cukup besar, terutama pada wilayah tertentu.

**Action:** Analisis ini dapat digunakan untuk mengidentifikasi wilayah dengan emisi tinggi dan populasi terdampak besar sehingga dapat menjadi prioritas dalam kebijakan pengurangan emisi, peningkatan mitigasi perubahan iklim, serta perlindungan terhadap masyarakat yang paling rentan terhadap dampaknya.


# **HeatMap**

```python
plt.figure(figsize=(10,8))
sns.heatmap(
    df[['temperature_change_c','co2_emissions_mt','sea_level_rise_mm',
        'heatwave_days','wildfire_incidents','climate_risk_score',
        'population_affected_m']].corr(),
    annot=True,
    cmap='viridis',
    fmt='.2f'
)
plt.title('Korelasi Antar Variabel Perubahan Iklim')
plt.show()
```

Warna pada heatmap menunjukkan kekuatan hubungan antar variabel sehingga memudahkan identifikasi pola hubungan dalam data.

**Insight:** Beberapa variabel menunjukkan korelasi yang lebih kuat dibandingkan yang lain, yang dapat mengindikasikan adanya keterkaitan antara faktor-faktor tertentu dalam fenomena perubahan iklim.

**Action:** Hasil korelasi ini dapat digunakan sebagai dasar untuk analisis lanjutan, seperti mengidentifikasi faktor yang paling berpengaruh terhadap risiko iklim atau perubahan suhu.
